```python
df.age = df.age.fillna(0)
df.cabin = df.cabin.fillna(0)
df.embarked = df.embarked.fillna(0)
```

```python
sns.heatmap(df.isnull())
```

```python
df = df.drop(columns = ['name','ticket','cabin'])
```

```python
df['sex'].replace( 'female', 0, inplace=True )
df['sex'].replace( 'male', 1, inplace=True )
df['Embarked_S'] = df['embarked'].replace('Q',0).replace('C',0).replace('S',1)
df['Embarked_Q'] = df['embarked'].replace('C',0).replace('S',0).replace('Q',1)
df['Embarked_C'] = df['embarked'].replace('S',0).replace('Q',0).replace('C',1)
```

```python
df
```

```python
x = df.iloc[:,0:12]
y = df['survived']
```

```python
x
```

```python
x= df.drop(columns= ['survived','embarked','boat','body','home.dest'])
x.fare = x.fare.fillna(0)
```

```python
x.isnull().sum()
```

```python
y
```

```python
x.head()
```

```python
num_cols = [0,1,4,3,6,7,8]
Kbin_cols = [5]
bina_cols= [2]
```

```python
from sklearn.preprocessing import StandardScaler,KBinsDiscretizer,Binarizer
from sklearn.compose import make_column_transformer
from sklearn import set_config
trans  = make_column_transformer((StandardScaler(),num_cols),
                                (KBinsDiscretizer(),Kbin_cols),
                                 (Binarizer(threshold = 50),bina_cols),
                                  remainder = 'passthrough')
set_config(display = 'diagram')
trans
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.34)  
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import make_pipeline
model = KNeighborsClassifier(9)
pipe = make_pipeline(trans,model)
pipe
```

```python
model
```

```python
pipe
```

```python
pipe.fit(x_train,y_train)
```

```python
pipe.fit(x_train,y_train)
```

```python
pred = pipe.predict(x_test)
pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(pred,y_test)*100
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
ConfusionMatrixDisplay.from_estimator(pipe,x_test,y_test)
```
